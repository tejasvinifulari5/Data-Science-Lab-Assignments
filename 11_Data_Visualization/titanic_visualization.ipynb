# 📊 Data Visualization - Titanic Dataset

## Objective
To analyze the Titanic dataset using visualization techniques and identify patterns related to survival, age, gender, and ticket fare.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.ffill(inplace=True)
df.bfill(inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Age", y="Sex", hue="Survived", data=df)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

plt.show()
```

```python
sns.histplot(data=df, x="Age", hue="Survived", multiple="stack")
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.pairplot(df, hue="Survived")
plt.show()
```

```python
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df)
plt.title("Survival Rate by Passenger Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
```

```python
sns.countplot(x="Embarked", hue="Survived", data=df)
plt.title("Survival Count by Embarkation Port")
plt.xlabel("Embarkation Port")
plt.ylabel("Count")
```

```python
sns.violinplot(x="Pclass", y="Age", hue="Survived", data=df)
plt.title("Age Distribution by Passenger Class and Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
```

```python
sns.stripplot(x="Pclass", y="Age", hue="Survived", data=df)
plt.title("Age Distribution by Passenger Class and Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
```

```python
sns.swarmplot(x="Pclass", y="Age", hue="Survived", data=df)
plt.title("Age Distribution by Passenger Class and Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
```

```python
print(df['Age'].describe())
```

```python
df.groupby('Survived')['Age'].mean()
```

```python
df.groupby(['Sex', 'Survived'])['Age'].mean()
```

```python
df['Survived'].value_counts()
```

```python
corr = df[['Age','Fare','Pclass','Survived']].corr() 
sns.heatmap(corr, annot=True) 
plt.title('Correlation Heatmap') 
plt.show() 
```

### 🔹 Observations

- Fare distribution is highly skewed (few passengers paid very high fares)  
- Females have higher survival rate than males  
- Younger passengers had better survival chances  
- First-class passengers generally survived more than third-class  

## ✅ Conclusion

Data visualization helps in understanding patterns and relationships in data.  
The Titanic dataset reveals that gender, age, and class significantly influenced survival chances.
