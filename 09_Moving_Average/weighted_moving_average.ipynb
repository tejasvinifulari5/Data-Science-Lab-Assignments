# 📊 Weighted Moving Average - Reliance Stock

## Objective
To implement Weighted Moving Average (WMA) on stock price data to give more importance to recent observations and analyze trends.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Reliance.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

data = df['Close']
```

```python
weights = np.array([1, 2, 3, 4, 5])
```

```python
def weighted_moving_average(series, weights):
    return series.rolling(len(weights)).apply(lambda x: np.dot(x, weights)/weights.sum(), raw=True)

df['WMA_5'] = weighted_moving_average(data, weights)
```

```python
plt.figure(figsize=(12,6))

plt.plot(data, label='Original Price')
plt.plot(df['WMA_5'], label='Weighted Moving Average (5 days)')

plt.title("Weighted Moving Average - Reliance Stock")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

### 🔹 Observations

- Weighted Moving Average gives more importance to recent data points  
- It reacts faster than Simple Moving Average  
- Helps in identifying short-term trends more effectively  

## ✅ Conclusion

Weighted Moving Average improves trend analysis by assigning higher importance to recent observations, making it more responsive to price changes.
