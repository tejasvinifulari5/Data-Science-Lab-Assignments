# 📊 Descriptive Statistics on Iris Dataset

## Objective
To compute statistical measures such as mean, median, standard deviation, and percentiles for different species in the Iris dataset and analyze their differences.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

The dataset contains measurements of iris flowers including sepal length, sepal width, petal length, petal width, and species.

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### 🔹 Observation
- Dataset contains 150 samples and 5 columns  
- Includes 3 species: Iris-setosa, Iris-versicolor, Iris-virginica  

```python
grouped = df.groupby('Species')
```

```python
grouped.mean()
```

```python
grouped.median()
```

```python
grouped.std()
```

```python
grouped.describe()
```

```python
np.percentile(df['SepalLengthCm'], [25, 50, 75])
```

```python
setosa_list = df[df['Species'] == 'Iris-setosa']['SepalLengthCm'].tolist()
versicolor_list = df[df['Species'] == 'Iris-versicolor']['SepalLengthCm'].tolist()
virginica_list = df[df['Species'] == 'Iris-virginica']['SepalLengthCm'].tolist()

print("Setosa:", setosa_list[:5])
print("Versicolor:", versicolor_list[:5])
print("Virginica:", virginica_list[:5])
```

### 🔹 Observations

- Iris-setosa has smaller petal length and width compared to other species  
- Iris-virginica generally has the largest measurements  
- Iris-versicolor lies between setosa and virginica  
- Mean and median values help understand central tendency  
- Standard deviation shows variability within each species  

## ✅ Conclusion

Descriptive statistics help summarize and understand the distribution of data.  
Grouping by species allows comparison between different categories effectively.
